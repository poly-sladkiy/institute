```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin
```

```python
a1 = 0.2
a2 = -1.4
a3 = 2
b1 = 4.001
b2 = -2.004
b3 = -3.6
```

```python
def func(x):
    return a1 * sin(b1 * x) + a2 * sin(b2 * x) + a3 * sin(b3 * x)
```

```python
x_start = -3.14
x_end = 3.14
x_delta = 0.01
```

```python
x = list(np.arange(x_start, x_end, x_delta))
```

```python
y = []
for i in x:
    y.append(func(i))
```

```python
for i, j in zip(x, y):
    print(f"{i:^3.4f} | {j:^3.4f}")
```

```python
plt.plot(x, y, 'g')
plt.grid()
```
